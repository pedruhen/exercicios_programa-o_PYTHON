```python
#QUESTAO 13)Elabore um algoritmo que leia um nmero de entrada que indicar a quantidade de
#registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
#pessoas e ao final apresentar a mdia de idade de ambos os gneros catalogados
soma_idade_masculina = 0
total_homens = 0
soma_idade_feminina = 0
total_mulheres = 0
N = int(input("Quantas pessoas voc vai registrar? "))

print(" Digite os dados de cada pessoa ")

for i in range(N):
    sexo = input(f"Sexo da {i+1} pessoa (M/F): ")
    idade = int(input(f"Idade da {i+1} pessoa: "))

    if sexo == 'M':
        soma_idade_masculina = soma_idade_masculina + idade
        total_homens = total_homens + 1
    else:
        soma_idade_feminina = soma_idade_feminina + idade
        total_mulheres = total_mulheres + 1

print("Resultados")


if total_homens > 0:
    media_masculina = soma_idade_masculina / total_homens
    print(f"Mdia de idade masculina: {media_masculina:.2f}")
else:
    print("Nenhum homem foi registrado.")


if total_mulheres > 0:
    media_feminina = soma_idade_feminina / total_mulheres
    print(f"Mdia de idade feminina: {media_feminina:.2f}")
else:
    print("Nenhuma mulher foi registrada.")
```

```python
#QUESTAO 16)Escreva um programa que v solicitando as idades dos alunos da sala at que todos
#sejam informados (perguntar ao usurio se deseja informar a idade do prximo aluno). Ao
#final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos tm mais de
#18 anos, quantos alunos tm at 18 anos, a mdia aritmtica e a mediana.
lista_idades = []
soma_idades = 0
resposta = 'S'
contador_mais_18 = 0


print("Informe Idades dos Alunos")
while resposta == 'S':
    idade = int(input("Digite a idade do aluno: "))


    lista_idades.append(idade)
    soma_idades = soma_idades + idade


    if idade > 18:
        contador_mais_18 = contador_mais_18 + 1

    resposta = input("Deseja informar a idade do prximo aluno? (S/N): ").upper()


print("\n--- Relatrio Final da Turma ---")


if len(lista_idades) > 0:

    mais_novo = min(lista_idades)
    mais_velho = max(lista_idades)


    total_alunos = len(lista_idades)


    ate_18 = total_alunos - contador_mais_18


    media = soma_idades / total_alunos


    lista_idades.sort()
    meio = len(lista_idades) // 2
    mediana = lista_idades[meio]

    print(f"Idade do aluno mais novo: {mais_novo} anos")
    print(f"Idade do aluno mais velho: {mais_velho} anos")
    print(f"Total de alunos com mais de 18 anos: {contador_mais_18}")
    print(f"Total de alunos com at 18 anos: {ate_18}")
    print(f"Mdia de idade da turma: {media:.2f} anos")
    print(f"Mediana das idades da turma: {mediana} anos")
else:
    print("Nenhuma idade foi informada.")
```
